```python
def scraping_albums(lyrics):
    """not today"""

songurls = []
album = []

for i in lyrics:
    r = requests.get(i, headers=headers)
    html = r.text
    print(r)
    mysoup = BeautifulSoup(html, 'html.parser')
    
    for i in mysoup.find_all('table', attrs={'class': 'table tdata'})[0].find_all('a'):
        url = i.get('href')
        if '/lyric/' in url:
            album.append(url)
             
    for i in album:
        songurls.append('https://www.lyrics.com'+i)
        
for i in songurls:
    var = requests.get(i, headers=headers)
    html_lyrics = var.text
    mysoup_lyrics = BeautifulSoup(html_lyrics, 'html.parser')
    k = mysoup_lyrics.find_all('pre')[0].text
    open(f'./data/lyrics.txt', 'a').write(k)
    
```

```python
lyrics_file_bbb = open('./data/lyrics_bbb.txt','r').read()
```

```python
lyrics_file_bbb = lyrics_file_bbb.splitlines()
```

```python
clean_lyrics_bbb = []
pattern = r"[^a-z-A-Z0-9]"

for i in lyrics_file_bbb:
    clean = re.sub(pattern, ' ', i.lower())
    clean_lyrics_bbb.append(clean)
```
